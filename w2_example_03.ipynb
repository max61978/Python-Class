````python
# def sub-routine
def GetOpinionData(opinionFile):
    print('即將要打開', opinionFile)
    ret_lines = ""
    with open(opinionFile, 'r') as read_file:
        lines = read_file.readlines()
    for line in lines:
        ret_lines += line

    print('Note ==> 檔案已經讀完')
    return ret_lines
````

````python
# 先來統計一下資料
# 準備詞典，不同的 call name 可視為指涉同一個SNS
# 字典型態複習 -- dict_obj = { k1:v1, k2:v2, ...}
dict_SNS = {
    'Facebook': ['facebook', 'fb', '臉書'],
    'Instagram': ['instagram', 'ig'],
    'Line': ['line'],
    'Twitter': ['twitter'],
    'YouTube': ['youtube'],
    '巴哈姆特': ['巴哈姆特'],
    'Dcard': ['dcard'],
    'Google': ['google'],
    'PTT': ['ptt'],
    'Moodle': ['moodle'],
    'Medium': ['medium'],
    'LetterBoxed': ['letterboxed'],
}

#print(dict_SNS)

# 用前面寫好的函式取得檔案內容
str_opinion = GetOpinionData('collected_opinion.txt')
str_opinion = str_opinion.lower()
#print(str_opinion)

#找意見表中每個terms的出現次數 -- using str.count()
dict_statSNS = {}

for k_sns in dict_SNS:
    #print(k_sns, '=>', dict_SNS[k_sns])
    count_k_sns = 0
    for callname in dict_SNS[k_sns]:
        #print('\t', callname)
        #print(str_opinion.count(callname))
        count_k_sns += str_opinion.count(callname)

    print(k_sns, '出現了', count_k_sns)
    dict_statSNS[k_sns] = count_k_sns

print(dict_statSNS)
````

````python
import matplotlib.pyplot as plt  #畫圖表模組


#準備畫圓餅圖

plt.figure(figsize=(18, 9))

y = dict_statSNS.keys()    # 顯示在圓餅圖上的 SNS 名稱
#size = len(dict_statSNS)  # 要顯示幾個？全部都顯示
x = dict_statSNS.values()   # 對應的資料數值

my_colors = ['red', 'orange', 'yellow', 'green', 'blue', 'darkblue', 'purple']  # 改改看顏色？

plt.pie(x,                              # 數值 x
        labels = y,                     # 標籤 y
        autopct = "%1.1f%%",            # 將數值百分比並留到小數點一位
        pctdistance = 0.6,              # 數字距離圓心的距離
        textprops = {"fontsize" : 12},  # 文字大小
        shadow=True)                    # 設定陰影

plt.axis('equal')                                          # 使圓餅圖比例相等
plt.title("本班同學參與SNS概況", {"fontsize" : 18})  # 設定標題及其文字大小
plt.legend(loc = "best")                                   # 設定圖例及其位置為最佳

plt.savefig("demo_stat_opinion.png")   # 儲存圖檔

plt.show()      # 顯示！
````
